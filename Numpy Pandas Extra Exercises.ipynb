```python
# Add 2 new columsn to the dataframe - first letter, length of word

first_letter = df['words'].astype(str).str[0]
length = df['words'].str.len()
df["length"] = length
df["firstLetter"] = first_letter
df
```

```python
# Use .groupby() and a builtin reduction/aggregation function to obtain the length of the longest word in each group of letters

grouped_df = df["length"].groupby(df["firstLetter"]).max()
grouped_df
```

```python
# Use .merge() to merge the original dataframe with the grouped dataframe on the column containing the first letter.

merged_df = df.merge(grouped_df, on = "firstLetter")
merged_df
```

```python
# Finally write code to pull out from the merged dataframe (use .loc[] or .iloc[]) 
# only those rows where the length column is equal to the maxlength column

longest_word_df = merged_df[merged_df["length_x"] == merged_df["length_y"]]
longest_word_df
```

### O - Produce an output that has three columns – letter, longest word, length. Sort in ascending order of the alphabet.

```python
output_df = longest_word_df[['firstLetter', 'words', 'length_y']]
output_df.columns = ['alphabet', 'longest word', 'length']
output_df = output_df.sort_values('alphabet')
output_df
```
